# Defination

# Bias

The error computation when we make model between predicted and actual value is called bias

# Variance

Means the spreadness of data or how our data is spread or randomly distributed between one and another

# Bias and Variance both are computed to be high and low

# Case I- When Our model has low bias and low variance

(1) Here Low Bias means the gap or difference between predicted and actual is very very low

(2) And Low Vaiance means the spreadness of data is also very low or they are scattered very less or they are present in group

# Case II-When our model has low bias and high variance 

(1) Here Low Bias means the gap or difference between predicted and actual is very very low or tends to be same

(2) And High varinace means the data points are scattered they are not in the from of group

# Case III- when our model has high bias and low variance

(1) Here High bias means there is huge or large gap or large error between actual value and predicted value

(2) And Low vaiance means they are not scattered or  in the form of group

# Case IV- when our model has high bias and high variance

(1) Here High bias means there is huge gap between predicted and actual

(2) And High variance means they are more scattered 

# Underfitting , Overfitting and Bestfit

# Overfitting

When our model is tries to fit or covers in all the data points plotted between x and y axis

# Underfitting

When our model is tries to not capture the underlying data

# Reasons of Overfitting

(1) They have very much knowledge of features of given datasets

(2) They usually gain too much trends of data and behavoiurs of data

(3) Train Dataset Error = 0  and Test Dataset Error = 100 or something  but Train is always = 0

# Reasons of Underfitting

(1) They have very less knowledge of features of given datasets

(2) They usually don't know the datasets trends and behavoiurs of data

# Realtions between bias, variance and underfitting,overfitting

# Overfitting Conditions

# Overfitting = High Variance Low Bias

# Underfitting Conditions

# Underfitting = Low Variance and High Biased

# In Simple words

(1)  Bias = Train Error

(2) Variance = Test Error

# Best Fit Model

A model is said to be best fit model which is low train error and low test error or in other words it has low bias and low variance is said to best fit or ideal model

# Bias Variance Tradeoff

When we plot prediction error rate with model complexity and try to get best fit model or low bias and low variance is model is called Bias variance Trade off

# Regularizations

Regularization is one of the most important concepts of machine learning. It is a technique to prevent the model from overfitting by adding extra information to it. Sometimes the machine learning model performs well with the training data but does not perform well with the test data.

(1) To reduce model complexity

(2) To reduce computational costs

i.e (i) y=0.9+1.3x+2.3x^2+3.4x^3

since in the given equation the x^3 coefficient has high computational cost 

If the above computational cost we perform for all records in the datasest then our model become more complex . To reduce this 

complexity of model so that the coefficients value get shrinks and we very easily performs computational we use regularizations

Now after shrinking the equation let say y=0.3+1.2x+2.0x^2+3.0x^3 

# Regularization Technique are 
# (1) Lasso Regressions (L1)

# Demonstration of Coefficients with Learning Rate

```python
import numpy as np
```

```python
X=np.array([1,2,4,3,5])
y=np.array([1,3,3,2,5])
```

```python
LR=0.1
m=5
c=0
epoch=0
```

```python
while epoch<1100:
    epoch+=1
    counter=0
    for i in X:
        yhat=(m*i)+c    #y=mx+c
        error=yhat-y[counter]     #error = y (predict)-y(counter)
        c=c-(LR*error)
        m=m-(LR*error*i)
        counter+=1
```

```python
print(m)
```

```python
print(c)
```

# Conclusion - Putting Differents LR(alpha) our coefficients also vary . Infact as Learning Rate is increases the coefficients also increases or comes bigger

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=load_boston()
```

```python
type(data)
```

```python
dataset = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
dataset.head(5)
```

```python
X=dataset.iloc[:,:-1]  ##Independent features
Y=dataset.iloc[:,-1] ##Dependent features
```

```python
Y.unique()
```

# Splitting Datasets into Train and Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
```

# Ridge Regressions

```python
from sklearn.linear_model import Ridge
```

# At alpha=1

```python
##Training the Model
ridge=Ridge(alpha=1)
ridge.fit(X_train,Y_train)
y_pred=ridge.predict(X_test)
```

```python
#Calculate mean square error 
mean_squared_error_ridge=np.mean((y_pred-Y_test)**2)
print(mean_squared_error_ridge)
```

```python
#Getting the ridge coefficients of all columns an print them
ridge_coefficient=pd.DataFrame()
ridge_coefficient['Columns']=X_train.columns
ridge_coefficient['Coefficient Estimate']=pd.Series(ridge.coef_)
print(ridge_coefficient)
#y=m1x1+m2x2+0
#|w|=|m1+m2+m3|
#|w|=|m1+m2|
```

# At alpha=1 the variabels are tends to be Zero but not zero

```python
#plotting this coefficients by visualizations
fig,ax=plt.subplots(figsize=(20,10))
ax.bar(ridge_coefficient['Columns'],
      ridge_coefficient['Coefficient Estimate'],color='Red')
ax.spines['bottom'].set_position('zero')
plt.style.use('ggplot')
plt.show()
```

# Here Computational Costly is CHAS,NOX,RM,DIS,PTRATIO,LSTAT

# At alpha=20

```python
##Training the Model
ridge=Ridge(alpha=20)
ridge.fit(X_train,Y_train)
y_pred=ridge.predict(X_test)
```

```python
#Calculate mean square error 
mean_squared_error_ridge=np.mean((y_pred-Y_test)**2)
print(mean_squared_error_ridge)
```

```python
#Getting the ridge coefficients of all columns an print them
ridge_coefficient=pd.DataFrame()
ridge_coefficient['Columns']=X_train.columns
ridge_coefficient['Coefficient Estimate']=pd.Series(ridge.coef_)
print(ridge_coefficient)
```

# At alpha=20 the variabels are tends to be Zero but the errro rate is decreased to 24 where as alpha = 1 error rate is 25

```python
#plotting this coefficients by visualizations
fig,ax=plt.subplots(figsize=(20,10))
ax.bar(ridge_coefficient['Columns'],
      ridge_coefficient['Coefficient Estimate'],color='Red')
ax.spines['bottom'].set_position('zero')
plt.style.use('ggplot')
plt.show()
```

# In Alpha=1   CHAS=2   but  Alpha = 20 CHAS=1.2 so at alpha=1 chas is more compuational costly
# Simliarly  At alpha=1 NOX=-10 but Alpha=20 NOX=-1 . so at alpha=1 NOX is more computational costly
# And we compare for all other columns

# Conclusion of Ridge Regressions

# (1) As we increase the alpha or learning rate our coefficient(m) of all columns decreases and tends to be zero not exact zero
# (2) As the coefficient(m) decreases the computational costs also decreases due to which we can say it's shrinkage the variable value
# (3) So those columns whose coefficients is near to zero remove that columns
# becuase it's totally reduce the calculation or we can say there is no calculations which will not affect our model accuracy infact it reduces error also
